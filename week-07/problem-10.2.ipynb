## (a)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.load('colquhounData.npy')
closed1 = data[:,[1]]
time = data[:,[0]]
closed = np.sum(closed1)
opened = np.cumsum(closed1)

opened_cumulative = 403 - opened[:,None]
```

## (b)

```python
plt.figure()
_=plt.plot(time,closed1,'.-')
plt.title('Original Data - Number of Closings')
_=plt.ylabel('Number of Closings')
_=plt.xlabel('Time')

plt.figure()
_=plt.plot(time,opened_cumulative,'.-')
plt.title('Number of Open Channels')
_=plt.ylabel('Number of Open Chanels')
_=plt.xlabel('Time')
```

The number of closed channels looks noisier.  I suppose this could be because the value of channels closing per time period can be completely random, while the number of channels remaining open will follow a curve in most instances.
