```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```python
def F(y,t):
    dy = [0,0]
    dy[0] = y[1]
    dy[1] = -y[0] + np.sin(0.8*t)
    return dy
```

```python
t = np.linspace(0,50,501)
y0 = [-3,8]
y = odeint(F,y0,t)
```

```python
print(t.shape, y.shape)
```

```python
plt.plot(t,y[:,0], 'r-')
plt.plot(t,y[:,1], 'b--')
```
