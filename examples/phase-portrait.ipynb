```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```python
t = np.linspace(-2,2,21)
X,Y = np.meshgrid(t,t)

U = -Y - 0.1 * X
V = X - 0.1 * Y

plt.figure(dpi=200)
plt.quiver(X,Y,U,V, pivot='mid')

def F(y,t):
    dy = [0,0]
    dy[0] = -y[1] - 0.1 * y[0]
    dy[1] = y[0] - 0.1 * y[1]
    return dy

t = np.linspace(0,40,1001)

# Solution 1
y0 = [-1.5,1]
y = odeint(F, y0, t)

plt.plot(y[:,0],y[:,1],'r-')
plt.axis('equal')

# Solution 2
y0 = [1,1.5]
y = odeint(F, y0, t)

plt.plot(y[:,0],y[:,1],'g-')
plt.axis('equal')
```
