```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1001)

firstOrder = 1/(1+x)
secondOrder = 1/(1+x**2)
tenthOrder = 1/(1+x**10)
```

```python
plt.figure(dpi=200)
plt.plot(x, firstOrder, label='N=1')
plt.plot(x, secondOrder, label='N=2')
plt.plot(x, tenthOrder, label='N=10')
plt.legend()
plt.loglog()
```
