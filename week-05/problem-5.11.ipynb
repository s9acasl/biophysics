## (a) Sums of random variables

Write Python code to generate a histogram of 50,000 samples of a random variable $y = x_1 + x_2$ where $x_1$ and $x_2$ are random variables drawn from the uniform distribution over $[-1/2, 1/2]$.  Superimpose a plot of the expected probability distribution function for this variable.

```python
# Standard setup
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
# Create a random number generator
from numpy.random import default_rng
rng = default_rng()

# Draw from the Uniform Distribution
rand = rng.random
```

```python
x1 = np.random.uniform(-.5,.5,50000)
x2 = np.random.uniform(-.5,.5,50000)
y = x1 + x2
#len(y)
_=plt.hist(y, bins = 40, rwidth = .7)
```

```python
_=sns.distplot(y, hist = True, kde = True, hist_kws = {'edgecolor':'black'}, kde_kws={'linewidth': 4})
_=plt.title("Histogram with Density Plot, 2 Random Variables")
```
