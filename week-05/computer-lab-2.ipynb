# 6.2

### a

```python
walks = 1
num_steps = 1000
x_final = []
y_final = []
displacement = []
while (walks <= 100):
    walks = walks+1
    x_step = 2*(rand(num_steps) < 0.5) - 1
    y_step = 2*(rand(num_steps) < 0.5) - 1
    x_step=np.cumsum(x_step)
    y_step=np.cumsum(y_step)
    
    x_final.append(x_step[-1]) 
    y_final.append(y_step[-1])
    displacement.append(np.sqrt(x_final[-1]**2 + y_final[-1]**2))
```

```python
_=plt.scatter(x_final,y_final)
_=plt.title("Scatter Plot of Final Values from 100 Random Walks")
```

### b

```python
_=plt.hist(displacement)
_=plt.title("Histogram of Displacement Values from 100 Random Walks")
```

### c

```python
displacement_squared=np.square(displacement)
_=plt.hist(displacement_squared)
_=plt.title("Displacement Squared Histogram")
```

### d

```python
hist, bins, _ = plt.hist(displacement_squared, bins=8)
logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
plt.subplot(211)
plt.hist(displacement_squared, bins=logbins)
plt.xscale('log')

plt.subplot(212)
_=plt.hist(displacement_squared, bins=logbins, log = True)
plt.xscale('log')
```

## a

```python
i = 0
#M = []
result = []
while i < 1000:
    i = i + 1
    n, p = 1, .08  # n = coins flipped, p = prob of success
    s = np.random.binomial(n, p, 1)
    result = np.append(result,s)
    #M = np.append(M,np.count_nonzero(s))
```

```python
x = np.nonzero(result)
x = np.diff(x)
x = x.flatten()
```

```python
_=plt.hist(x)
_=plt.title('Histogram of Waiting Times between Heads in 1000 Flips')
plt.xlabel('Waiting Time')
_=plt.ylabel('Frequency')
```

```python
_=plt.hist(x, log = True)
_=plt.title('Semilog Histogram of Waiting Times between Heads in 1000 Flips')
plt.xlabel('Waiting Time')
_=plt.ylabel('Frequency')
```

## b

```python
x1 = np.sum(x)
x2 = len(x)
Average = x1/x2
print("The Average wait time between heads is ", round(Average), "Seconds")
```
