## (a) __Histogram of $1/x$__

```python
# Standard setup
import numpy as np
import matplotlib.pyplot as plt

# Create a random number generator
from numpy.random import default_rng
rng = default_rng()

# Draw from the Uniform Distribution
rand = rng.random
```

```python
y = np.random.uniform(0,1,10000)
y = np.reciprocal(y)
#print(y)
```

```python
_=plt.hist(y, bins = 50, range = (1, 10))
_=plt.title("Reciprocal of 10000 Uniformly Distributed Values")
#len(bins)
```

## (b) Log-log plot of frequencies

```python
_=plt.hist(y, bins=50, log = True, range = (1,10))
_=plt.xscale('log')
_=plt.title("Log Log Histogram 10000 Values")
```
